```python
df = pd.read_csv('car_showroom_data.csv')
```

```python
df.head()
```

```python
df.describe(include="all")
```

```python
# ======================
# Data Analysis
```

```python
df.head()
```

```python
df.info()
```

```python
# Analysis 1 - Total Sales & Revenue - Overall business performance
total_sales = df.shape[0]
total_revenue = df['Price'].sum()
avg_price = df['Price'].mean()
```

```python
total_sales
```

```python
print(f'Total_Car_Sold is: {total_sales}')
print(f'Total_Revenue_Generated is: {total_revenue}')
print(f'Average_Selling_Price is: {avg_price}')
```

```python
# Analysis 2 - City-wise Sales Count - Which cities perform best
sales_by_city = (
    df.groupby('City').size().reset_index(name='Sales_Count').sort_values(by='Sales_Count', ascending=False)
)
```

```python
sales_by_city
```

```python
# Analysis 3 - City-wise Revenue - High volume  high revenue
city_revenue = (
    df.groupby('City')['Price'].sum().reset_index(name='Total_Revenue').sort_values(by='Total_Revenue', ascending=False)
)
```

```python
city_revenue
```

```python
# Analysis 4 - Average price by city - Customer purchasing power
avg_price_city = (
    df.groupby('City')['Price'].mean().reset_index(name='Average_Price').sort_values(by='Average_Price', ascending=False)
)
```

```python
avg_price_city
```

```python
# Analysis 5 - Most Sold Car Models - Product demand analysis
model_sales = (
    df.groupby('CarModel').size().reset_index(name='Units_Sold').sort_values(by='Units_Sold', ascending=False)
)
```

```python
model_sales
```

```python
# Analysis 6 - Revenue by Car Model - Which model earns most money
model_revenue = (
    df.groupby('CarModel')['Price'].sum().reset_index(name='Total_Revenue').sort_values(by='Total_Revenue', ascending=False)
)
```

```python
model_revenue
```

```python
# Analysis 7 - Fuel Type Distribution - Market trend detection
fuel_dist = (
    df['FuelType'].value_counts().reset_index().rename(columns={'index':'FuelType', 'FuelType':'Count'})
)
```

```python
fuel_dist
```

```python
# Analysis 8 - Fuel Type  Revenue - Which fuel earns more
fuel_revenue = (
    df.groupby('FuelType')['Price'].sum().reset_index(name='Revenue').sort_values(by='Revenue', ascending=False)
)
```

```python
fuel_revenue
```

```python
df.info()
```

```python
df['SaleDate'] = pd.to_datetime(df['SaleDate'])
df['SaleDate'] = df['SaleDate'].astype(object)
```

```python
# Analysis 9 - Year-wise / Date-wise Sales Trend - Time series insight
yearly_sales = (
    df.groupby('SaleYear').size().reset_index(name='Sales')
)
```

```python
yearly_sales
```

```python
# Analysis 10 - Salesperson Performance -  Employee productivity
```

```python
salesperson_perf = (
    df.groupby("SalesPersonID")
      .agg(
          Cars_Sold=("Price", "count"),
          Revenue=("Price", "sum")
      )
      .reset_index()
      .sort_values(by="Revenue", ascending=False).head(10)
)
```

```python
salesperson_perf
```

```python
# Analysis 11 - Customer-Level Analysis - Customer value
Customer_value = (
    df.groupby('CustomerID')['Price'].agg(['count','sum']).reset_index()
    .rename(columns={'count':'Cars_Bought','sum':'Total_Spend'})
)
Customer_value = Customer_value.sort_values(by='Total_Spend', ascending=False).head(10)
```

```python
Customer_value
```

```python
# Analysis 12 - Cross Analysis (City  Fuel Type) - Advanced business insight
city_fuel = pd.crosstab(df['City'], df['FuelType'])
```
